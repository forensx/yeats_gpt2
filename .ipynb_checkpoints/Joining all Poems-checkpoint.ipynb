# Joining all Poems

```python
import pandas as pd
import numpy as np
import os as os
```

```python
all_poems = os.listdir("Poems")
```

```python
full_db = []

for poem in all_poems:
    
    f = open("Poems/{}".format(poem))
    single_poem = f.read()
    
    full_db.append(single_poem)
    
    f.close()
```

```python
concat = []
fin = [concat.append(poem) for poem in full_db]
```

```python
concat
```

```python
f = open("FullPoemsDB.txt", "w+")
for poem in concat:
    f.write("{}\r".format(poem))
f.close()
```
